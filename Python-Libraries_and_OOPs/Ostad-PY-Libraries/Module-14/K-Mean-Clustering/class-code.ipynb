```python
recency = maxDate.drop('InvoiceDate', axis=1)
recency
```

```python
freq = df.drop_duplicates(subset="InvoiceNo")
freq
```

```python
freq = freq.groupby(["CustomerID"]).count()[['InvoiceNo']]
```

```python
freq
```

```python
df['total'] = df['Quantity'] * df["UnitPrice"]
df
```

```python
money = df.groupby('CustomerID')[["total"]].sum()
money
```

```python
RFM = pd.concat([recency, freq, money], axis=1)
RFM
```

```python
RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(RFM)
```

```python
scaled
```

```python
scaled.shape
```

```python
from sklearn.cluster import KMeans
```

<h2> To Find The Value of K </h2>

```python
k_range = range(1, 10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(scaled)
    sse.append(km.inertia_)
```

```python
sse, k_range
```

```python
plt.plot(k_range, sse, 'r-*')
plt.xlabel = 'k'
plt.ylabel = 'Sum of Squared Error'
```

```python
# object creation and data fitting
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
kmeans.labels_ + 1
```

```python
RFM['Cluster'] = (kmeans.labels_ + 1)
```

```python
RFM
```

```python
final = RFM.groupby('Cluster').mean()[['Recency', 'Frequency', 'Monetary']]
final
```

```python
def func(row):
    if row['Cluster'] == 3:
        return 'Diamond'
    elif row['Cluster'] == 2:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['group'] = RFM.apply(func, axis=1)
```

```python
RFM
```

```python
result = RFM.group.value_counts()
```
