```python
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets
ires = datasets.load_iris()
ires
```

```python
df = pd.DataFrame(data=ires.data, columns= ires.feature_names)
df['flower'] = ires.target
df
```

```python
X = df.drop('flower', axis=1)
y = df['flower']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=50)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=1, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

<h2 style="color:whitesmoke;"> Using K Fold Cross Validation </h2>

```python
from sklearn.model_selection import cross_val_score
```

```python
sv1 = cross_val_score(SVC(kernel='rbf', C=1, gamma='auto'), X, y, cv=5)
sv1.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
sv2 = cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
sv2.mean()
```

```python
from sklearn.naive_bayes import GaussianNB
sv3 = cross_val_score(GaussianNB(), X, y, cv=5)
sv3.mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier
sv4 = cross_val_score(KNeighborsClassifier(n_neighbors=3), X, y, cv=5)
sv4.mean()
```

<h2 style="color:green;"> Use GridSearchCV </h2>

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'), {
    'C': [1, 10, 20, 30],
    'kernel' : ['rbf', 'linear', 'poly'],
}, cv=5)
clf.fit(X, y)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
```
