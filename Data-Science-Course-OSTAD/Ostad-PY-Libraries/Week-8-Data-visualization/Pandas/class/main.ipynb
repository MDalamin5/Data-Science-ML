```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
gas['UK'].plot()
```

```python
gas.plot(x='Year', y='USA')
```

```python
gas.plot(x='Year', y='USA', xticks=gas.Year[::3].tolist(), figsize=(15, 5))
```

```python
plt.subplots(figsize=(15, 5))
plt.plot(gas['Year'], gas['USA'])
plt.plot(gas['Year'], gas['UK'])
plt.plot(gas['Year'], gas['Canada'])
plt.xticks(gas.Year[::3].tolist())
plt.show()
```

```python
plt.subplots(figsize=(15, 5))
plt.title('Gas Price Over the time(in USD)', fontdict={'fontweight' : 'bold', 'fontsize' : 18, 'color' : 'g'})
# plt.plot(gas['Year'], gas['USA'])
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas['Canada'], 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas['Australia'], 'y.-', label='Aus')
plt.xlabel('Dller')
plt.ylabel('Year')
plt.legend()
plt.xticks(gas.Year[::3].tolist())
plt.show()
plt.savefig('gas-price-analysis.png', dpi=300)
```
