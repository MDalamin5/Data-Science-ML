```python
fig1 = plt.figure(figsize=(8, 5))
# [left, bottom, width, height (range 0--1)]

axes = fig1.add_axes([.1, .1, .9, .9])
axes.plot(x, y, 'm--')

axes.set_xlabel('This is X-axis')
axes.set_title('This is new Title')

axes.set_ylim([0, 50])
# axes.set_xlim([1, 4])


fig1.savefig('my-first-graphDATA', dpi = 200)
```

```python
fig3 = plt.figure(figsize=(10, 5))

axes_main = fig3.add_axes([0.1, 0.1, 0.9, 0.9])

axes_inset = fig3.add_axes([.3, .5, .3, .4])

axes_main.plot(x, y, 'g')

axes_inset.plot(y, x, 'r')

fig3.savefig('alamin.pdf', dpi = 200)

# you can set level using previous method
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label = 'x**2')
axes.plot(x, x**3, label = 'x**3')

axes.legend(loc = 2)
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (10, 8))

axes.plot(x, x+1, linewidth = .24)
axes.plot(x, x+2, linewidth = .5)
axes.plot(x, x+3, linewidth = .64)
axes.plot(x, x+4, linewidth = .74)
axes.plot(x, x+5, linewidth = .84)
axes.plot(x, x+6, linewidth = .94)
axes.plot(x, x+7, linewidth = 1)
```
