```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
print(x, y)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r-*')
plt.show()
```

```python
plt.plot(x, y, 'r--')
plt.xlabel('This is X-axis Title')
plt.ylabel('This is Y-axis Title')
plt.title('Figure Ploat')
plt.xticks(range(0, 18, 3))
plt.show()
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'r-*')

plt.subplot(1, 3, 2)
plt.plot(x, y, 'g-*')

plt.subplot(1, 3, 3)
plt.plot(x, y, 'b-*')

plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(8, 5))
axes = fig1.add_axes([.1, .1, .9, .9])
axes.plot(x, y, 'r-o')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('This is most advanced Graph Dewawing')
fig1.savefig('test graph.png', dpi=200)
```

```python
print(x)
print(x**2)
print(x**3)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label = 'x**2')
axes.plot(x, x**3, label = 'x**3')

axes.legend(loc=2)
plt.show()
```

```python
fig, axes = plt.subplots()
axes.plot(x, x+1, 'r.-')
axes.plot(x, x+2, 'g--')
```

```python
fig, axes = plt.subplots(figsize=(10, 8))
axes.plot(x, x+1, color='red', linewidth=0.25)
axes.plot(x, x+2, color='green', linewidth=0.50)
axes.plot(x, x+3, color='black', linewidth=0.76)
axes.plot(x, x+4, color='blue', linewidth=1)

axes.plot(x, x+5, color='green', lw=5, linestyle='-')
axes.plot(x, x+6, color='green', lw=5, linestyle='-.')
axes.plot(x, x+7, color='green', lw=5, linestyle=':')

axes.plot(x, x+8, color='black', lw=3, ls='-', marker='+')
axes.plot(x, x+9, color='black', lw=3, ls='--', marker='o')
axes.plot(x, x+10, color='black', lw=3, ls='-.', marker='s')
axes.plot(x, x+11, color='black', lw=3, ls=':', marker='1', markersize=20)

axes.plot(x, x+12, color='green', lw=3, ls='-', marker='o', markersize=2)
axes.plot(x, x+13, color='green', lw=3, ls='-', marker='o', markersize=4)
axes.plot(x, x+14, color='green', lw=3, ls='-', marker='o', markersize=8, markerfacecolor='red')
axes.plot(x, x+15, color='green', lw=3, ls='-', marker='s', markersize=16)
```
