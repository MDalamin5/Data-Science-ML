## Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x = ['python', 'c', 'c++', 'java']
y = [85, 70, 60, 82]

c = ['black', 'y', 'g', 'b']
plt.title('Most Populer Language')
plt.xlabel('Language', fontsize = 15)
plt.ylabel("No.", fontsize = 14)
plt.bar(x, y, color = c, width= 0.5, edgecolor = 'r', label = "Test-Label")
plt.legend()
plt.show()
```
