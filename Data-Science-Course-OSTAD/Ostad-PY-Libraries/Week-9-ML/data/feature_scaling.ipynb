```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data_set = pd.read_csv('feature scaling.csv')
data_set
```

```python
x = data_set[['Age', 'Salary']]
x
```

```python
from sklearn.preprocessing import MinMaxScaler
MXS = MinMaxScaler()
x_after_min_max_scaler = MXS.fit_transform(x)
x_after_min_max_scaler
```

```python
# Standardisation
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
x_after_standardisation = std.fit_transform(x)
x_after_standardisation
```

```python
X_train_scaled = pd.DataFrame(x_after_standardisation, columns=x.columns)
X_train_scaled
```
